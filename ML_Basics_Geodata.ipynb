```python
from zipfile import ZipFile

with ZipFile('/content/Communes_MGP.shp.zip','r') as zipO:

  zipO.extractall(path='/content/')
```

```python
import geopandas as gpd

shapefile = gpd.read_file('/content/Communes_MGP.shp')

print(shapefile)
```

```python
shapefile.plot()
```

```python
import fiona
```

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

kmlfile = gpd.read_file("Communes_MGP.kml")

print(kmlfile)
```

```python
kmlfile.plot()
```

```python
geojsonfile = gpd.read_file("Communes_MGP.json")
print(geojsonfile)
```

```python
geojsonfile.plot()
```
