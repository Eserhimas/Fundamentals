```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
path = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

df = pd.read_csv(path)
df.head()
```

```python
# drop irrelevant variables

df.drop(['PassengerId','Pclass', 'Name', 'Cabin', 'Embarked', 'Ticket'], inplace=True, axis=1)
df.head()
```

```python
# Encoding

sex = pd.get_dummies(df['Sex'], drop_first=True)
sex.head()
```

```python
df_s = pd.concat([df, sex], axis=1)

df_s.head()
```

```python
df_s.drop(['Sex', 'Fare'], axis=1, inplace=True)
df_s.head()
```

```python
X = df_s.iloc[:,[1,2,3,4]]
y = df_s.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.20, random_state=42)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# missing values imputing

df_s.isnull().sum() # Age

val = df.Age.mean()

X_train.Age = X_train.Age.fillna(val)
X_test.Age = X_test.Age.fillna(val)
```

```python
X_test.isnull().sum()
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

```

```python
y_hat = lr.predict(X_test)
y_hat
```

```python
cm = confusion_matrix(y_test, y_hat)
cm
```

```python
print(classification_report(y_test, y_hat))
```
